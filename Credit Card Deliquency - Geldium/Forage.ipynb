#### Handle Missing Data

```python
# Identify all columns with missing data
print("üîç Columns with missing values:")
missing_summary = df.isnull().sum()
missing_columns = missing_summary[missing_summary > 0]
print(missing_columns)

# Create a table to track how we'll handle missing data
missing_data_table = pd.DataFrame({
    'Column': missing_columns.index,
    'Missing_Count': missing_columns.values,
    'Percentage_Missing': (missing_columns.values / len(df) * 100).round(2)
})

print("\nüìä Missing Data Summary:")
display(missing_data_table)

# Let's decide how to handle each missing column
print("\nüí° Proposed handling strategy:")

# For Income - fill with median (middle value)
if 'Income' in df.columns and df['Income'].isnull().sum() > 0:
    income_median = df['Income'].median()
    df['Income'] = df['Income'].fillna(income_median)
    print(f"‚úÖ Income: Filled {df['Income'].isnull().sum()} missing values with median: {income_median}")

# For Loan_Balance - fill with median
if 'Loan_Balance' in df.columns and df['Loan_Balance'].isnull().sum() > 0:
    loan_median = df['Loan_Balance'].median()
    df['Loan_Balance'] = df['Loan_Balance'].fillna(loan_median)
    print(f"‚úÖ Loan_Balance: Filled {df['Loan_Balance'].isnull().sum()} missing values with median: {loan_median}")

# For Credit_Score - we'll check if there are any missing
if 'Credit_Score' in df.columns and df['Credit_Score'].isnull().sum() > 0:
    credit_median = df['Credit_Score'].median()
    df['Credit_Score'] = df['Credit_Score'].fillna(credit_median)
    print(f"‚úÖ Credit_Score: Filled {df['Credit_Score'].isnull().sum()} missing values with median: {credit_median}")

print("\nüéØ Missing values after handling:")
print(f"Remaining missing values: {df.isnull().sum().sum()}")
```

#### Identify Risk Factors

```python
# Let's create a simple "risk score" based on payment history
def count_late_missed_payments(row):
    """Count how many times a customer was late or missed payments"""
    payment_columns = ['Month_1', 'Month_2', 'Month_3', 'Month_4', 'Month_5', 'Month_6']
    count = 0
    for col in payment_columns:
        if row[col] in ['Late', 'Missed']:
            count += 1
    return count

# Apply this function to each row
df['Late_Missed_Count'] = df.apply(count_late_missed_payments, axis=1)

print("üö® Risk Analysis - Late/Missed Payments:")
print(f"Average late/missed payments: {df['Late_Missed_Count'].mean():.2f}")
print(f"Maximum late/missed payments: {df['Late_Missed_Count'].max()}")
print(f"Customers with 3+ late/missed payments: {(df['Late_Missed_Count'] >= 3).sum()}")

# Analyze high-risk customers
high_risk = df[df['Late_Missed_Count'] >= 4]
print(f"\nüî¥ High-risk customers (4+ late/missed payments): {len(high_risk)}")

# Compare high-risk vs low-risk customers
low_risk = df[df['Late_Missed_Count'] <= 1]

print("\nüìä Comparison - High Risk vs Low Risk Customers:")
print(f"High Risk - Average Credit Score: {high_risk['Credit_Score'].mean():.2f}")
print(f"Low Risk - Average Credit Score: {low_risk['Credit_Score'].mean():.2f}")

print(f"\nHigh Risk - Average Credit Utilization: {high_risk['Credit_Utilization'].mean():.3f}")
print(f"Low Risk - Average Credit Utilization: {low_risk['Credit_Utilization'].mean():.3f}")
```

## DATA PREPARATION & FEATURE SELECTION

### Comprehensive Feature Analysis

```python
# Feature Analysis and Selection
print("üîç PHASE 1: FEATURE ANALYSIS & SELECTION")
print("=" * 60)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest, f_classif

# Define target variable
df['High_Risk_Customer'] = (df['Late_Missed_Count'] >= 3).astype(int)

# Handle missing values
df['Income'] = df['Income'].fillna(df['Income'].median())
df['Loan_Balance'] = df['Loan_Balance'].fillna(df['Loan_Balance'].median())

# Encode employment status
employment_mapping = {
    'Employed': 2, 'EMP': 2, 'employed': 2, 'Self-employed': 1,
    'Unemployed': 0, 'retired': 1, 'Retired': 1
}
df['Employment_Status_Encoded'] = df['Employment_Status'].map(employment_mapping).fillna(1)

print("‚úÖ Data prepared with target variable")

# Feature correlation analysis
feature_candidates = [
    'Age', 'Income', 'Credit_Score', 'Credit_Utilization', 
    'Missed_Payments', 'Loan_Balance', 'Debt_to_Income_Ratio',
    'Account_Tenure', 'Employment_Status_Encoded'
]

X = df[feature_candidates]
y = df['High_Risk_Customer']

# Feature importance using SelectKBest
selector = SelectKBest(score_func=f_classif, k='all')
selector.fit(X, y)

feature_scores = pd.DataFrame({
    'Feature': feature_candidates,
    'F_Score': selector.scores_,
    'P_Value': selector.pvalues_
}).sort_values('F_Score', ascending=False)

print("\nüìä FEATURE IMPORTANCE SCORES:")
display(feature_scores)

# Select top 5 features
top_5_features = feature_scores.head(5)['Feature'].tolist()
print(f"\nüéØ TOP 5 FEATURES SELECTED: {top_5_features}")

# Correlation heatmap
plt.figure(figsize=(12, 8))
correlation_matrix = X.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('Feature Correlation Matrix')
plt.tight_layout()
plt.show()
```

### Data Pipeline Setup

```python
# Data Pipeline Setup
print("\nüîß DATA PIPELINE SETUP")
print("=" * 50)

# Use top 5 features for our models
final_features = top_5_features
X_final = df[final_features]
y_final = df['High_Risk_Customer']

print(f"Final feature set: {final_features}")

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_final, y_final, test_size=0.2, random_state=42, stratify=y_final
)

# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print(f"Training set: {X_train.shape[0]} samples")
print(f"Testing set: {X_test.shape[0]} samples")
print(f"High-risk rate in training: {y_train.mean():.3f}")
print(f"High-risk rate in testing: {y_test.mean():.3f}")

# Save feature names and scaler for deployment
import joblib
joblib.dump(scaler, 'feature_scaler.pkl')
print("‚úÖ Data pipeline prepared and scaler saved")
```

## MODEL DEVELOPMENT & COMPARISON

### Implement Both Model Options

```python
# Implement Simple and Complex Models
print("\nü§ñ PHASE 2: MODEL DEVELOPMENT")
print("=" * 60)

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
import time

# MODEL 1: Simple - Logistic Regression
print("üìà MODEL 1: LOGISTIC REGRESSION (Simple)")
start_time = time.time()

lr_model = LogisticRegression(
    random_state=42,
    class_weight='balanced',
    max_iter=1000,
    C=1.0  # Regularization strength
)

lr_model.fit(X_train_scaled, y_train)
y_pred_lr = lr_model.predict(X_test_scaled)
y_proba_lr = lr_model.predict_proba(X_test_scaled)[:, 1]

lr_time = time.time() - start_time

# MODEL 2: Complex - Random Forest
print("üå≥ MODEL 2: RANDOM FOREST (Complex)")
start_time = time.time()

rf_model = RandomForestClassifier(
    n_estimators=100,
    max_depth=8,
    min_samples_split=20,
    min_samples_leaf=10,
    class_weight='balanced',
    random_state=42
)

rf_model.fit(X_train, y_train)  # No scaling needed for Random Forest
y_pred_rf = rf_model.predict(X_test)
y_proba_rf = rf_model.predict_proba(X_test)[:, 1]

rf_time = time.time() - start_time

print("‚úÖ Both models trained successfully")
```

### Comprehensive Model Comparison

```python
# Model Performance Comparison
print("\nüìä COMPREHENSIVE MODEL COMPARISON")
print("=" * 60)

def evaluate_model(name, y_true, y_pred, y_proba, train_time):
    """Evaluate model performance comprehensively"""
    return {
        'Model': name,
        'Accuracy': accuracy_score(y_true, y_pred),
        'Precision': precision_score(y_true, y_pred),
        'Recall': recall_score(y_true, y_pred),
        'F1_Score': f1_score(y_true, y_pred),
        'AUC_Score': roc_auc_score(y_true, y_proba),
        'Training_Time': train_time
    }

# Evaluate both models
lr_metrics = evaluate_model('Logistic Regression', y_test, y_pred_lr, y_proba_lr, lr_time)
rf_metrics = evaluate_model('Random Forest', y_test, y_pred_rf, y_proba_rf, rf_time)

# Create comparison table
comparison_df = pd.DataFrame([lr_metrics, rf_metrics])
comparison_df = comparison_df.round(4)
comparison_df['Training_Time'] = comparison_df['Training_Time'].round(3)

print("PERFORMANCE COMPARISON:")
display(comparison_df)

# Feature importance analysis
print("\nüîç FEATURE IMPORTANCE ANALYSIS:")

# Logistic Regression coefficients
lr_importance = pd.DataFrame({
    'Feature': final_features,
    'Coefficient': lr_model.coef_[0],
    'Absolute_Impact': abs(lr_model.coef_[0])
}).sort_values('Absolute_Impact', ascending=False)

print("Logistic Regression Feature Impact:")
display(lr_importance)

# Random Forest feature importance
rf_importance = pd.DataFrame({
    'Feature': final_features,
    'Importance': rf_model.feature_importances_
}).sort_values('Importance', ascending=False)

print("Random Forest Feature Importance:")
display(rf_importance)

# Visualize feature importance comparison
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Logistic Regression coefficients
ax1.barh(lr_importance['Feature'], lr_importance['Coefficient'], color='skyblue')
ax1.set_title('Logistic Regression: Feature Coefficients')
ax1.axvline(x=0, color='black', linestyle='-', alpha=0.3)

# Random Forest importance
ax2.barh(rf_importance['Feature'], rf_importance['Importance'], color='lightgreen')
ax2.set_title('Random Forest: Feature Importance')

plt.tight_layout()
plt.show()
```

```python
# Comprehensive Fairness Assessment
print("\n‚öñÔ∏è COMPREHENSIVE FAIRNESS ASSESSMENT")
print("=" * 60)

# Create test set with predictions and ALL necessary demographic columns
# First, let's make sure we have all the original columns we need
test_results = X_test.copy()

# Add back the original columns we need for fairness analysis
original_columns_needed = ['Age', 'Income', 'Credit_Score', 'Missed_Payments', 
                          'Credit_Utilization', 'Debt_to_Income_Ratio']

# Check which columns are missing and add them from the original dataframe
for col in original_columns_needed:
    if col not in test_results.columns:
        # Get the indices from our test set and fetch the original values
        test_indices = X_test.index
        test_results[col] = df.loc[test_indices, col]

# Add prediction results
test_results['Actual_Risk'] = y_test.values
test_results['Predicted_Risk'] = y_pred_selected
test_results['Prediction_Probability'] = y_proba_selected
test_results['Correct_Prediction'] = (y_test.values == y_pred_selected)

print("‚úÖ Test dataset prepared with all necessary columns")
print(f"Columns in test_results: {test_results.columns.tolist()}")

# Define demographic groups for fairness analysis
print("\nüìä Defining demographic groups for fairness analysis...")

# Age groups
test_results['Age_Group'] = pd.cut(test_results['Age'], 
                                   bins=[18, 30, 45, 60, 100], 
                                   labels=['18-30', '31-45', '46-60', '60+'])

# Income groups - handle any missing values
if 'Income' in test_results.columns:
    test_results['Income_Group'] = pd.cut(test_results['Income'],
                                          bins=[0, 30000, 60000, 100000, float('inf')],
                                          labels=['Low', 'Medium', 'High', 'Very High'])
    print("‚úÖ Income groups created")
else:
    print("‚ö†Ô∏è Income column not available for grouping")

# Credit score groups
if 'Credit_Score' in test_results.columns:
    test_results['Credit_Score_Group'] = pd.cut(test_results['Credit_Score'],
                                               bins=[300, 580, 670, 740, 800, 900],
                                               labels=['Poor', 'Fair', 'Good', 'Very Good', 'Excellent'])
    print("‚úÖ Credit score groups created")
else:
    print("‚ö†Ô∏è Credit_Score column not available for grouping")

# Payment behavior groups
if 'Missed_Payments' in test_results.columns:
    test_results['Payment_Behavior_Group'] = pd.cut(test_results['Missed_Payments'],
                                                   bins=[-1, 0, 2, 10],
                                                   labels=['No Missed', '1-2 Missed', '3+ Missed'])
    print("‚úÖ Payment behavior groups created")

def calculate_fairness_metrics(group_df, group_name):
    """Calculate comprehensive fairness metrics for a group"""
    if len(group_df) == 0:
        return None
    
    try:
        metrics = {
            'Group': group_name,
            'Sample_Size': len(group_df),
            'Actual_Risk_Rate': group_df['Actual_Risk'].mean(),
            'Predicted_Risk_Rate': group_df['Predicted_Risk'].mean(),
            'Selection_Rate': group_df['Predicted_Risk'].mean(),
            'Accuracy': group_df['Correct_Prediction'].mean(),
            'False_Positive_Rate': ((group_df['Actual_Risk'] == 0) & (group_df['Predicted_Risk'] == 1)).sum() / max(1, (group_df['Actual_Risk'] == 0).sum()),
            'False_Negative_Rate': ((group_df['Actual_Risk'] == 1) & (group_df['Predicted_Risk'] == 0)).sum() / max(1, (group_df['Actual_Risk'] == 1).sum()),
        }
        
        # Calculate precision and recall with error handling
        if (group_df['Predicted_Risk'] == 1).sum() > 0:
            metrics['Precision'] = precision_score(group_df['Actual_Risk'], group_df['Predicted_Risk'], zero_division=0)
        else:
            metrics['Precision'] = 0
            
        if (group_df['Actual_Risk'] == 1).sum() > 0:
            metrics['Recall'] = recall_score(group_df['Actual_Risk'], group_df['Predicted_Risk'], zero_division=0)
        else:
            metrics['Recall'] = 0
            
        return metrics
    except Exception as e:
        print(f"Error calculating metrics for {group_name}: {e}")
        return None

# Calculate fairness metrics for all groups
fairness_results = []

print("\nüìà Calculating fairness metrics...")

# Age groups
if 'Age_Group' in test_results.columns:
    for age_group in test_results['Age_Group'].cat.categories:
        group_data = test_results[test_results['Age_Group'] == age_group]
        metrics = calculate_fairness_metrics(group_data, f"Age_{age_group}")
        if metrics:
            fairness_results.append(metrics)

# Income groups
if 'Income_Group' in test_results.columns:
    for income_group in test_results['Income_Group'].cat.categories:
        group_data = test_results[test_results['Income_Group'] == income_group]
        metrics = calculate_fairness_metrics(group_data, f"Income_{income_group}")
        if metrics:
            fairness_results.append(metrics)

# Credit score groups
if 'Credit_Score_Group' in test_results.columns:
    for credit_group in test_results['Credit_Score_Group'].cat.categories:
        group_data = test_results[test_results['Credit_Score_Group'] == credit_group]
        metrics = calculate_fairness_metrics(group_data, f"Credit_{credit_group}")
        if metrics:
            fairness_results.append(metrics)

# Payment behavior groups
if 'Payment_Behavior_Group' in test_results.columns:
    for payment_group in test_results['Payment_Behavior_Group'].cat.categories:
        group_data = test_results[test_results['Payment_Behavior_Group'] == payment_group]
        metrics = calculate_fairness_metrics(group_data, f"Payment_{payment_group}")
        if metrics:
            fairness_results.append(metrics)

# Create fairness report
if fairness_results:
    fairness_df = pd.DataFrame(fairness_results)
    print("\nFAIRNESS ASSESSMENT ACROSS CUSTOMER GROUPS:")
    display(fairness_df.round(3))

    # Identify potential bias issues
    print("\nüîç BIAS DETECTION ANALYSIS:")
    print("=" * 40)

    overall_selection_rate = test_results['Predicted_Risk'].mean()
    overall_fpr = ((test_results['Actual_Risk'] == 0) & (test_results['Predicted_Risk'] == 1)).sum() / max(1, (test_results['Actual_Risk'] == 0).sum())
    overall_accuracy = test_results['Correct_Prediction'].mean()

    print(f"Overall Selection Rate: {overall_selection_rate:.3f}")
    print(f"Overall False Positive Rate: {overall_fpr:.3f}")
    print(f"Overall Accuracy: {overall_accuracy:.3f}")

    # Check for disparate impact (80% rule)
    print("\nüö® POTENTIAL DISPARATE IMPACT (Selection Rate < 80% of Overall):")
    disparate_impact_found = False
    for _, row in fairness_df.iterrows():
        if overall_selection_rate > 0:  # Avoid division by zero
            selection_ratio = row['Selection_Rate'] / overall_selection_rate
            if selection_ratio < 0.8:
                print(f"  ‚Ä¢ {row['Group']}: ratio = {selection_ratio:.3f}")
                disparate_impact_found = True
    
    if not disparate_impact_found:
        print("  ‚úÖ No disparate impact detected")

    # Check for accuracy disparities
    print("\n‚ö†Ô∏è ACCURACY DISPARITIES (>10% difference from overall):")
    accuracy_issues_found = False
    for _, row in fairness_df.iterrows():
        accuracy_diff = abs(row['Accuracy'] - overall_accuracy)
        if accuracy_diff > 0.1:
            print(f"  ‚Ä¢ {row['Group']}: accuracy = {row['Accuracy']:.3f} (diff: {accuracy_diff:.3f})")
            accuracy_issues_found = True
    
    if not accuracy_issues_found:
        print("  ‚úÖ No significant accuracy disparities")

    # Check for high false positive rates
    print("\nüî¥ HIGH FALSE POSITIVE RATES (> Overall + 0.1):")
    fpr_issues_found = False
    for _, row in fairness_df.iterrows():
        if row['False_Positive_Rate'] > overall_fpr + 0.1:
            print(f"  ‚Ä¢ {row['Group']}: FPR = {row['False_Positive_Rate']:.3f}")
            fpr_issues_found = True
    
    if not fpr_issues_found:
        print("  ‚úÖ No high false positive rate issues")

else:
    print("‚ùå No fairness results could be calculated. Check available columns.")
    print(f"Available columns: {test_results.columns.tolist()}")

# Visualize fairness metrics
if fairness_results:
    print("\nüìä VISUALIZING FAIRNESS METRICS...")
    
    # Create subplots for different fairness metrics
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))
    
    # Plot 1: Selection Rates by Group
    if len(fairness_df) > 0:
        axes[0,0].barh(fairness_df['Group'], fairness_df['Selection_Rate'], color='lightblue')
        axes[0,0].axvline(x=overall_selection_rate, color='red', linestyle='--', label='Overall Rate')
        axes[0,0].set_title('Selection Rates by Customer Group')
        axes[0,0].set_xlabel('Selection Rate (Predicted High Risk)')
        axes[0,0].legend()
    
    # Plot 2: Accuracy by Group
    if len(fairness_df) > 0:
        axes[0,1].barh(fairness_df['Group'], fairness_df['Accuracy'], color='lightgreen')
        axes[0,1].axvline(x=overall_accuracy, color='red', linestyle='--', label='Overall Accuracy')
        axes[0,1].set_title('Accuracy by Customer Group')
        axes[0,1].set_xlabel('Accuracy')
        axes[0,1].legend()
    
    # Plot 3: False Positive Rates by Group
    if len(fairness_df) > 0:
        axes[1,0].barh(fairness_df['Group'], fairness_df['False_Positive_Rate'], color='salmon')
        axes[1,0].axvline(x=overall_fpr, color='red', linestyle='--', label='Overall FPR')
        axes[1,0].set_title('False Positive Rates by Customer Group')
        axes[1,0].set_xlabel('False Positive Rate')
        axes[1,0].legend()
    
    # Plot 4: Sample Sizes by Group
    if len(fairness_df) > 0:
        axes[1,1].barh(fairness_df['Group'], fairness_df['Sample_Size'], color='gold')
        axes[1,1].set_title('Sample Sizes by Customer Group')
        axes[1,1].set_xlabel('Number of Customers')
    
    plt.tight_layout()
    plt.show()

print("\n‚úÖ Fairness assessment completed!")
```

INPUT: Customer Features ‚Üí [Feature Scaling] ‚Üí [Logistic Function] ‚Üí [Probability Score] ‚Üí [Threshold Comparison] ‚Üí OUTPUT: Risk Classification

STEP-BY-STEP PROCESS:
1. Data Ingestion: Collect customer financial data (income, credit score, payment history, etc.)
2. Feature Scaling: Normalize numerical features to equal scale using StandardScaler
3. Linear Combination: Calculate weighted sum of features: z = Œ≤‚ÇÄ + Œ≤‚ÇÅx‚ÇÅ + Œ≤‚ÇÇx‚ÇÇ + ... + Œ≤‚Çôx‚Çô
4. Probability Conversion: Apply sigmoid function: P(HighRisk) = 1 / (1 + e^(-z))
5. Classification: Compare probability to threshold (default: 0.5)
   - If P(HighRisk) ‚â• 0.5 ‚Üí Predict "High Risk"
   - If P(HighRisk) < 0.5 ‚Üí Predict "Low Risk"

KEY COMPONENTS:
- Coefficients (Œ≤): Learned weights indicating feature importance
- Intercept (Œ≤‚ÇÄ): Baseline risk level
- Sigmoid Function: Converts linear output to probability (0 to 1)
- Regularization: Prevents overfitting through parameter C

MATHEMATICAL FOUNDATION:
P(Y=1|X) = 1 / (1 + e^(-(Œ≤‚ÇÄ + Œ≤‚ÇÅx‚ÇÅ + Œ≤‚ÇÇx‚ÇÇ + ... + Œ≤‚Çôx‚Çô)))

WHERE:
- P(Y=1|X) = Probability of delinquency given features X
- Œ≤·µ¢ = Coefficient indicating feature i's impact on risk
- x·µ¢ = Value of feature i for a specific customer
